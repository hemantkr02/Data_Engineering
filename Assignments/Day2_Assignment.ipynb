```python
2. Where to use Map, filter and reduce?
```

```python
#Map()

it is used to transform data and when we want to apply some operation to every element
```

```python
num=[1,2,3,4]
result = list(map(lambda x:x*x, num))
print (result)
```

```python
# Filter
It is used when we want to select elements based upon condition
```

```python
num=[1,2,3,4]
result = list(filter(lambda x:x%2 == 0, num))
print (result)
```

```python
# reduce()
it is used when we want to reduce list into single value
```

```python
from functools import reduce
num=[1,2,3,4]
result = reduce(lambda x,y:x*y, num)
print (result)
```
