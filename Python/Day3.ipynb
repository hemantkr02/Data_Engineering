```python
class TestAdd(unittest.TestCase): #This defines a test case class using Pythonâ€™s built-in unittest framework
    def test_add_positive(self):
        self.assertEqual(add(2, 3), 5) #assertEqual checks if the result of add(2, 3) is equal to 5

    def test_add_negative(self):
        self.assertEqual(add(-1, -1), -2)

    def test_add_zero(self):
        self.assertEqual(add(0, 5), 5)

#When you run this file with python,
#Python will automatically look for classes that inherit from unittest.TestCase and run all methods that start with test_
```
