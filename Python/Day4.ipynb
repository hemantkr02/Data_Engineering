```python
def factorial(n):
    if not isinstance(n, int):
        raise TypeError("Input must be an integer")
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
import unittest

def factorial(n):
    if not isinstance(n, int):
        raise TypeError("Input must be an integer")
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

class TestFactorial(unittest.TestCase):

    def test_positive(self):
        self.assertEqual(factorial(6), 720)

    def test_zero(self):
        self.assertEqual(factorial(0), 1)

    def test_negative(self):
        with self.assertRaises(ValueError):
            factorial(-4)

    def test_string(self):
        with self.assertRaises(TypeError):
            factorial("a")

#unittest.main()
suite = unittest.TestLoader().loadTestsFromTestCase(TestFactorial)
unittest.TextTestRunner().run(suite)
```

```python
import unittest

def factorial(num):
    if not isinstance(num, int):
        return "not a integer"
    if num < 0:
        return "negative number"
    elif num == 0 or num == 1:
        return 1
    else:
        return num * factorial(num - 1)

class TestFactorial(unittest.TestCase):

    def test_positive_number(self):
        self.assertEqual(factorial(2), 2, "failed for positive number")

    def test_negative_number(self):
        self.assertEqual(factorial(-4), "negative number", "failed for negative numbers")

    def test_zero(self):
        self.assertEqual(factorial(0), 1, "failed for 0")

    def test_string_input(self):
        self.assertEqual(factorial("s"), "not a integer", "failed for string")


#unittest.TestFactorial()
suite = unittest.TestLoader().loadTestsFromTestCase(TestFactorial)
unittest.TextTestRunner().run(suite)
```

```python
#Iterators: it is an object that allows to traverse through all the elements of a collection like tuple of lists one at a time
#Implements _iter_() and _next_() methods
#Raises StopIteration when no more data is available
```

```python
class Mynumber:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        if self.num<=5:
            x=self.num
            self.num += 1
            return x
        else:
            raise StopIteration
myclass = Mynumber()
myiter = iter(myclass)

for x in myiter:
    print(x)
```

```python
#Generator: it is a special type of iterator that is defined using a function and the 'yield' keyword
# yeild is used insted of yeild
# Generates value one at a time
```

```python

def my_generator():
    for i in range(1,6):
        yield i
gen = my_generator()
for value in gen:
    print(value)
```

```python
# Iterator vs Generator:

#Iterator : Requires class with __iter__() and __next__()
#Gen       : uses function with yield keyword

#Iterator: MOre code , manual implementation
#Genarator: simple and concise implementation

#iterator : more control
#generator: more memory efficient
```

```python
class multiple:
    def __iter__(self):
        self.num=5
        return self
    def __next__(self):
        if self.num <= 50:
            x = self.num
            self.num+=5
            return x
        else:
            raise StopIteration
myclass = multiple()
myiter = iter(myclass)

for x in myiter:
    print(x)
```

```python
def even (num1,num2):
    for i in range(num1,num2+1):
        if (i%2==0):
            yield i

gen = even(10,20)
for value in gen:
    print(value)
```

```python
class ShowCart:
    def __iter__(self):
        self.cart = ['milk', 'bread', 'biscuit']
        self.index = 0
        return self

    def __next__(self):
        if self.index < len(self.cart):
            item = self.cart[self.index]
            self.index += 1
            return item
        else:
            raise StopIteration

myclass = ShowCart()
myiter = iter(myclass)

for x in myiter:
    print(x)
```
