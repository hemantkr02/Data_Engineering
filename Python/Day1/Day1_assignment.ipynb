```python
# Assignment1: To do the transpose using function and without function
```

```python
# Without Function
matrix = [
    [1,2,3,4],
    [6,7,8,9],
    [7,8,9,5]
]
transpose =[]
for i in range(len(matrix[0])):
    row = []
    for j in range(len(matrix)):
        row.append(matrix[j][i])
    transpose.append(row)
print(transpose)
```

```python
#With function
def transpose_matrix(mat):
    res=[]
    for i in range(len(mat[0])):
        row=[]
        for j in range(len(mat)):
            row.append(mat[j][i])
        res.append(row)
    return res

matrix = [
    [1,2,3,4],
    [5,6,7,8],
    [7,8,9,2]
]

print(transpose_matrix(matrix))
```
