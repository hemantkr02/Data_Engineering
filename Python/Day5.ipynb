```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
traffic = [150, 2, 180, 220, 250, 310]
#create line plot
plt.plot(months, traffic, marker='^', markersize = 10, color = 'blue', linestyle='-.', linewidth=3)

#add labels
plt.xlabel('Month')
plt.ylabel('Monthly Traffic(in thousands)')
plt.title('Monthly Website Traffic')

#Display
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
product_a_revenue = [45, 55, 40, 70, 80, 75]
product_b_revenue = [35,40, 50, 55, 70, 68]

plt.plot(months, product_a_revenue, marker='o',color = 'blue', linestyle='-.', label='Product A')

plt.plot(months, product_b_revenue, marker='o',color = 'red', linestyle='-.', label='Product A')

#add labels
plt.xlabel('Month')
plt.ylabel('Monthly revenue(in thousands)')
plt.title('Monthly revenue comparison')

#diaplay a legend to differenciate between A and B
plt.legend()
#Display
plt.show()
```

```python
import matplotlib.pyplot as plt

# Matches
matches = ['Match 1', 'Match 2', 'Match 3', 'Match 4', 'Match 5', 'Match 6']

# Runs scored by Player A and Player B
player_a_runs = [45, 60, 30, 75, 90, 55]
player_b_runs = [50, 40, 65, 70, 85, 60]

# Plot Player A
plt.plot(matches, player_a_runs, marker='+', markersize='10', color='green', linestyle='-.', label='Player A')

# Plot Player B
plt.plot(matches, player_b_runs, marker='+',markersize='10',  color='red', linestyle='-.', label='Player B')

# Add labels
plt.xlabel('Match')
plt.ylabel('Runs Scored')
plt.title('Cricket Performance Comparison')

# Display legend
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count matches per season
matches_per_season = df['season'].value_counts()

# Plot
plt.figure(figsize=(10,6))
matches_per_season.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of IPL Matches per Season (2008–2020)')
plt.xlabel('season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count number of wins per team
wins_per_team = df['winner'].value_counts()

# Plot
plt.figure(figsize=(12,6))
wins_per_team.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Total IPL Wins by Team (2008–2020)')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.show()
```
