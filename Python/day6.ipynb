```python
import matplotlib.pyplot as plt

quarters=['Q1','Q2','Q3','Q4']
electronics=[10000,12000,11000,10500]
clothing=[5000,6000,7500,8000]
homeappliance=[3000,7500,8200,9000]

plt.figure(figsize=(10,6))
plt.title("Retail sales")
plt.xlabel("quaterly data")
plt.ylabel("sales")
plt.stackplot(quarters,electronics,clothing,homeappliance,labels=['electronics','clothing','homeappliance'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)

x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2 +y**2))

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c=z,cmap='viridis')
ax.set_xlabel("X_axiz")
ax.set_ylabel("y_axiz")
ax.set_zlabel("z_axiz")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(-x)
fig,axs=plt.subplots(2,2,figsize=(10,8))

axs[0,0].plot(x,y1,'r')
axs[0,0].set_title('sine_Wave')
axs[0,1].plot(x,y2,'b')
axs[0,1].set_title('cose_Wave')
axs[1,0].plot(x,y3,'g')
axs[1,0].set_title('tangent_Wave')
axs[1,1].plot(x,y4,'m')
axs[1,1].set_title('exponent_decay')

plt.tight_layout()
plt.show()
```
